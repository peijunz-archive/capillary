# Measurement of the Motion

```python
from capillary import edge, analyse
```

Extract all the information of the 73 frames

```python
rho, theta, phi1, phi2=analyse.analyse(edge.get, range(1, 74));
```

```python
def r2d(r):
    '''Radian to Degree'''
    return r*180/pi
```

```python
plot(r2d(theta), 'o', markersize=3)
xlabel('Frame')
ylabel(r'$\theta/^\circ$')
title('Displacement Angle');
```

```python
plot(rho, 'o-', markersize=3, label=r"$R$")
title('Displacement')
xlabel('Frame')
ylabel(r'$\rho/$px');
```

```python
plot(r2d(phi1), 'o-', markersize=2, label=r"$\phi_1$");
plot(r2d(phi2), 'o-', markersize=2, label=r"$\phi_2$");
legend(loc='upper left');
xlabel('Frame')
ylabel(r'$\phi/^\circ$')
title('Orientation');
```

```python
plot(r2d(phi1+phi2)/2, 'o', markersize=3);
xlabel('Frame');
title(r"$(\phi_1+\phi_2)/2$");
```
