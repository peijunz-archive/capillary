```python
rho, theta, phi1, phi2, fp, fn=analyse.analyse(edge.G[0], range(1, 2900+1), cachefile='data/1.npy')
```

```python
def r2d(r):
    '''Radian to Degree'''
    return r*180/pi
```

```python
plot(r2d(theta)[-100:], 'o', markersize=1)
#plot(10+r2d(theta), 'o', markersize=1)
xlabel('Frame')
ylabel(r'$\theta/^\circ$')
title('Displacement Angle');
```

```python
plot(rho, 'o', markersize=2, label=r"$R$")
title('Displacement')
xlabel('Frame')
ylabel(r'$\rho/$px');
```

```python
from scipy.signal import medfilt
plot(r2d((phi1)), 'o', markersize=1, label=r"$\phi_1$");
plot(r2d((phi2)), 'o', markersize=1, label=r"$\phi_2$");
legend(loc='upper left');
xlabel('Frame')
ylabel(r'$\phi/^\circ$')
title('Orientation');
```

```python
plot(r2d(phi2)+60+r2d(theta), 'o', markersize=1);
plot(r2d(phi1)+r2d(theta)+240, 'o', markersize=1);
xlabel('Frame');
title(r"$(\phi_1+\phi_2)/2$");
```
